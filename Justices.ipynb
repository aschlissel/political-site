```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import warnings
warnings.filterwarnings('ignore')
from datetime import datetime, date
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_Justices_of_the_Supreme_Court_of_the_United_States'
soup = BeautifulSoup(requests.get(url).text)
```

```python
print(soup.prettify())
```

```python
table = soup.find('table', {'class': 'wikitable sortable'})
```

```python
#table
```

```python
len(table.find_all('tr'))
```

```python
justice = []
state = []
position = []
replacing = []
date_confirmed = []
tenure = []
tenure_length = []
appointed_by = []

for row in table.find_all('tr')[1:118]:
    col = row.find_all('td')
    #print()
    #print(col)
    #print()
    justice.append(col[1].text.replace('\n', ' '))
    state.append(col[2].text.replace('\n', ' '))
    position.append(col[3].text.replace('\n', ' '))
    replacing.append(col[4].text.replace('\n', ' '))
    date_confirmed.append(col[5].text.replace('\n', ' '))
    tenure.append(col[6].text.replace('\n', ' '))
    tenure_length.append(col[7].text.replace('\n', ' '))
    try:
        appointed_by.append(col[8].text)
        appointed_by_prev = col[8].text
    except:
        appointed_by.append(appointed_by_prev)
```

```python
assert len(justice) == len(state) == len(position) == len(replacing) == len(date_confirmed) == len(tenure) == len(tenure_length) == len(appointed_by)
```

```python
starts = []
ends = []
```

```python
for s in tenure:
    div = s.index('–')
    start = s[5:div].replace('[b]', '').replace('[c]', '').strip()
    try:
        end = s[div+2:s.index('(')].strip()
    except:
        end = s[div+2:].strip()
    start = datetime.strptime(start, '%B %d, %Y').date()
    if end.lower()=="incumbent":
        end = date.today()
    else:
        end = datetime.strptime(end, '%B %d, %Y').date()
    print("start: %s" %start)
    print("end: %s" %end)
    starts.append(start)
    ends.append(start)
    
```

```python
df = pd.DataFrame({'justice': justice, 'state': state, 'position': position, 'replacing': replacing, 'date_confirmed': date_confirmed,
      'tenure_length': tenure_length, 'appointed_by': appointed_by, 'start': starts, 'end': ends})
```
